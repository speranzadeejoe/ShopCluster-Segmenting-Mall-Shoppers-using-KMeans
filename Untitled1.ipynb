```python
import sklearn
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
# 📦 Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X2 = df[['Annual Income (k$)', 'Spending Score (1-100)', 'Gender']]
X2_scaled = StandardScaler().fit_transform(X2)

kmeans2 = KMeans(n_clusters=5, random_state=42)
labels2 = kmeans2.fit_predict(X2_scaled)
df['Cluster_with_gender'] = labels2

print("Silhouette Score (with gender):", silhouette_score(X2_scaled, labels2))

# Visualize
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster_with_gender', palette='Set1')
plt.title("Clusters With Gender")
plt.show()
```

```python
# Select features
X1 = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Scale the features
scaler1 = StandardScaler()
X1_scaled = scaler1.fit_transform(X1)

# Apply KMeans
kmeans1 = KMeans(n_clusters=5, random_state=42)
labels1 = kmeans1.fit_predict(X1_scaled)
df['Cluster_wo_gender'] = labels1

# Evaluate
score1 = silhouette_score(X1_scaled, labels1)
print(f'Silhouette Score (without gender): {score1:.2f}')

# Visualize
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster_wo_gender', palette='Set2')
plt.title("Clusters Without Gender")
plt.show()
```
